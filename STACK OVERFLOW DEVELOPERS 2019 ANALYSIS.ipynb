```python
#DROPPING THOSE COLUMNS WHICH WILL NOT BE USED IN OUR ANALYSIS
data.drop(columns={'Respondent','OpenSource','Student','EdLevel','EduOther','OrgSize','YearsCode','Age1stCode','MgrIdiot','MgrMoney','MgrWant','JobSeek','LastHireDate','LastInt','JobFactors','ResumeUpdate','CurrencyDesc','CompTotal','CompFreq','ConvertedComp','WorkRemote','ImpSyn','UnitTests','PurchaseHow','PurchaseWhat','OffOn','SocialMedia','Extraversion','ScreenName','SOVisit1st','SOVisitFreq','SOVisitTo','SOFindAnswer','SOTimeSaved','SOHowMuchTime','SOAccount','SOPartFreq','SOJobs','EntTeams','SOComm','SONewContent','Gender','Trans','Sexuality','Ethnicity','Dependents','SurveyLength','SurveyEase','FizzBuzz','CurrencySymbol','WelcomeChange','WorkPlan','WorkChallenge'},inplace=True)
```

```python
#AFTER DROPPING THE COLUMNS
data.columns
```

```python
#ANALYSIS ON MAIN BRANCH
data['MainBranch'].head(50)
```

```python
#REPLACING THE CONTENTS OF MAIN BRANCH WITH ONE WORDS
data['MainBranch'].replace({'I am a student who is learning to code':'Student','I am not primarily a developer, but I write code sometimes as part of my work':'Not a Developer','I am a developer by profession':'Developer','I code primarily as a hobby':'Hobby of Coding','I used to be a developer by profession, but no longer am':'Used to be a Developer'},inplace=True)
```

```python
#CHECKING IT IS REPLACED OR NOT
data['MainBranch'].head(50)
```

```python
#PLOTTING TO SEE WHICH OF THE BRANCH HAS MORE NUMBER OF PEOPLE IN MAIN BRANCH
plt.figure(figsize=(20,5))
sns.countplot(data['MainBranch'],data=data)
plt.xlabel('Branch Type', fontsize=10)
plt.ylabel('Count')
plt.title("Main Branch")
plt.savefig('Main Branch.png',dpi=300)
plt.show()
```

```python
#AS YOU CAN SEE THAT THE "DEVELOPERS" ARE THE MOST IN THE STACK OVERFLOW DEVELOPERS SURVEY IN 2019 
```

```python
#marking job satifaction on a scale of 0 to 4, 2 being neutral
data_developer['JobSat'].replace({'Slightly satisfied':1, 'Very satisfied':4, 'Very dissatisfied':0,
       'Slightly dissatisfied':3, 'Neither satisfied nor dissatisfied':2},inplace=True)
```

```python
#ANALYSIS ON COUNTRY AND JOB SATISFACTION
Job_sat=data_developer.groupby('Country')['JobSat'].sum().sort_values(ascending=False).head().reset_index()
Job_sat.plot.bar(x='Country',y='JobSat')
plt.xlabel('Country',fontsize=10)
plt.ylabel("Job Satisfaction Rating Sum",fontsize=10)
plt.show()
```

```python
#AS YOU CAN SEE THAT ACCORDING TO JOB SATISFACTION UNITED STATES HAS THE MAXIMUM NO OF SATISFIED DEVELOPERS IN 2019
```

```python
#ANALYSIS ON WORK OF HOURS PER WEEK ACCORDING TO COUNTRY WISE
work_per_week=data_developer.groupby('Country')['WorkWeekHrs'].sum().sort_values(ascending=False).head().reset_index()
work_per_week.plot.bar(x='Country',y='WorkWeekHrs')
plt.xlabel('COUNTRY',fontsize=10)
plt.ylabel("WORK PER WEEK SUM",fontsize=10)
plt.show()
```
